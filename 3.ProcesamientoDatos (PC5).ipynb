```python
#Procesamiento reporte
#Una empresa quiere contratar a 5 nuevos trabajadores por eso necesita que le proporcionenen
#el correo electronico de las 5 personas con mayor experiencia (YOE) y mayor puntaje (Code challenge score).
#Ademas quiere saber la nacionalidad (country) y la fecha(aplication date)en que tomo la prueba. 
```

```python
#primero ordenar los datos por experiencia (YOE) y puntaje (Code Challenge Score) de forma descendente
df_sorted = df.sort_values(by=['YOE', 'Code Challenge Score'], ascending=[False, False])

#correos electrónicos de las 5 personas con mayor experiencia y mayor puntaje
top_5_emails = df_sorted.head(5)['Email'].tolist()
top_5_names_emails = df_sorted.head(5)[['First Name', 'Last Name', 'Email']]

print("Nombres y correos electrónicos de los mas aptos:")
print(top_5_names_emails)

#nacionalidad (country) y la fecha (Application Date) de las 5 personas seleccionadas
top_5_nationalities = df_sorted.head(5)[['Country', 'Application Date']]

print("\nNacionalidad y Fecha de prueba aplicada de los mas aptos:")
print(top_5_nationalities)
```
